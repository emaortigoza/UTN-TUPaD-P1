# ‚ú® Verificaci√≥n de N√∫meros Perfectos

En este notebook veremos c√≥mo determinar si un n√∫mero natural es **perfecto**.

Un n√∫mero perfecto es aquel que es igual a la suma de sus divisores propios positivos (excluyendo el mismo n√∫mero).

Ejemplos cl√°sicos:
- 6 ‚Üí 1 + 2 + 3 = 6 ‚úÖ
- 28 ‚Üí 1 + 2 + 4 + 7 + 14 = 28 ‚úÖ

## üî¢ Funci√≥n `leer_entero_validado()`

```python
def leer_entero_validado(mensaje, min = float("-Inf"), max = float("Inf")):
    n = int(input(f"{mensaje}: "))
    while n < min or n > max:
        n = int(input(f"ERROR. {mensaje}: "))
    return n
```

üîÅ Esta funci√≥n pide al usuario un n√∫mero entero y repite el pedido hasta que est√© dentro del rango permitido.

## üîÅ Funci√≥n `obtener_resto()` (sin `%`)

```python
def obtener_resto(num1, num2):
    return num1 - num2 * (num1 // num2)
```

‚úèÔ∏è Calcula el resto de una divisi√≥n sin usar el operador `%`.

## ‚úÖ Funci√≥n `es_multiplo()`

```python
def es_multiplo(x, y):
    return obtener_resto(x, y) == 0
```

üìå Determina si `x` es m√∫ltiplo de `y` usando `obtener_resto()`.

## ‚ûï Funci√≥n `sumatoria_divisores_propios()`

```python
def sumatoria_divisores_propios(numero):
    sumatoria = 0
    for i in range(1, numero // 2 + 1):
        if es_multiplo(numero, i):
            sumatoria += i
    return sumatoria
```

üßÆ Recorre los divisores propios del n√∫mero (hasta la mitad) y los suma si son divisores.

## ‚ú® Funci√≥n `es_perfecto()`

```python
def es_perfecto(numero):
    # Algunos nums perfectos: 6, 28, 496, 8128
    return sumatoria_divisores_propios(numero) == numero
```

üîç Compara la suma de los divisores propios con el n√∫mero original para saber si es perfecto.

## üì¢ Funci√≥n `informar_si_numero_es_perfecto()`

```python
def informar_si_numero_es_perfecto(numero):
    if es_perfecto(numero):
        print(f"El numero {numero} es perfecto")
    else:
        print(f"El numero {numero} NO es perfecto")
```

üì£ Muestra un mensaje indicando si el n√∫mero ingresado es perfecto o no.

## üß™ Programa principal

```python
num = leer_entero_validado("Ingrese un n√∫mero natural", 1)
informar_si_numero_es_perfecto(num)
```

‚úÖ El usuario ingresa un n√∫mero natural y el programa informa si es un n√∫mero perfecto.
